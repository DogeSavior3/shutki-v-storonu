```python
import re
import os
```

```python
def clean_text(text):
    pattern = r'[^ \n\u0590-\u05FF.,!?;:"\']'
    cleaned_text = re.sub(pattern, '', text)
    
    cleaned_text = re.sub(r'[ ]{2,}', ' ', cleaned_text)
    cleaned_text = re.sub(r'\n{3,}', '\n\n', cleaned_text)
    cleaned_text = re.sub(r'\n \n', '\n', cleaned_text)

    return cleaned_text.replace('\n\n', '\n')
    
def read_text_from_file(file_path):
    with open(file_path, 'r') as file:
        return file.read()

def write_text_to_file(text, file_path):
    with open(file_path, 'w') as file:
        file.write(text)
```

```python
write_text_to_file(clean_text(read_text_from_file('תקוה ופחד.txt')).replace('\n\n', '\n'),'תקוה ופחד.txt')
```

```python
def process_texts(input_folder, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for file_name in os.listdir(input_folder):
        input_file_path = os.path.join(input_folder, file_name)

        if os.path.isfile(input_file_path):
            print(f"Processing file: {file_name}")

            try:
                text = read_text_from_file(input_file_path)
            except:
                continue

            cleaned_text = clean_text(text)
            output_file_path = os.path.join(output_folder, file_name)
            write_text_to_file(cleaned_text, output_file_path)

            print(f"Finished processing file: {file_name}")
```

```python
input_folder = 'pure'
output_folder = 'cleaned'

process_texts(input_folder, output_folder)
```

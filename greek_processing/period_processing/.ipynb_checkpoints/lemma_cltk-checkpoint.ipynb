```python
import os
from cltk import NLP
from greek_accentuation.characters import base

input_folder = 'processed'
output_folder = 'lemmatized_cltk'

if not os.path.exists(output_folder):
    os.makedirs(output_folder)

cltk_nlp = NLP(language="grc")
cltk_nlp.max_length = 10000000000

def clean_text(text):
    cleaned_text = ''.join(base(char) for char in text)
    return cleaned_text
```

```python
def split_text(text, chunk_size=50000):
    return [text[i:i + chunk_size] for i in range(0, len(text), chunk_size)]

# Обработка файлов
for filename in os.listdir(input_folder):
    if filename.endswith('.txt'):
        file_path = os.path.join(input_folder, filename)

        with open(file_path, 'r', encoding='utf-8') as file:
            text = file.read()

        cleaned_text = clean_text(text)
        chunks = split_text(cleaned_text)

        lemmata = []
        for chunk in chunks:
            cltk_doc = cltk_nlp.analyze(text=chunk)
            lemmata.extend(cltk_doc.lemmata)

        output_file_path = os.path.join(output_folder, filename)
        with open(output_file_path, 'w', encoding='utf-8') as output_file:
            output_file.write("\n".join(lemmata))

        print(f"Лемматизация завершена для файла: {filename}")

print("Лемматизация завершена для всех файлов.")
```

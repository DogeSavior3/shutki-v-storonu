```python
import PyPDF2

def extract_text_from_pdf(pdf_path):
    pdf_file = open(pdf_path, 'rb')
    pdf_reader = PyPDF2.PdfReader(pdf_file)
    
    text = ""
    for page_num in range(len(pdf_reader.pages)):
        page = pdf_reader.pages[page_num]
        text += page.extract_text()
    
    pdf_file.close()
    return text

def write_text_to_file(text, txt_path):
    with open(txt_path, 'w') as txt_file:
        txt_file.write(text)
```

```python
def clean_text(text):
    pattern = r'[^ \n\u0370-\u03FF\u1F00-\u1FFF]'
    cleaned_text = re.sub(pattern, '', text)
    cleaned_text = cleaned_text.replace('  ','').replace('  ','').replace('  ','').replace('\n \n','\n').replace('\n \n','\n').replace('\n \n \n','\n\n').replace('\n\n\n','\n\n').replace('\n\n\n','\n\n').replace('\n\n\n','\n\n').replace('\n \n\n','\n\n').replace('\n\n \n','\n\n')
    return cleaned_text

def read_text_from_file(file_path):
    with open(file_path, 'r') as file:
        return file.read()

def write_text_to_file(text, file_path):
    with open(file_path, 'w') as file:
        file.write(text)
```

```python
pdf_path = '01_Ilias.pdf'
txt_path = '01_Ilias.txt'

pdf_text = extract_text_from_pdf(pdf_path)

write_text_to_file(pdf_text, txt_path)

print(f"Text extracted from {pdf_path} and saved to {txt_path}")
```

```python
def replace_in_file(file_path, old_string, new_string):
    with open(file_path, 'r') as file:
        file_contents = file.read()
        
    updated_contents = file_contents.replace(old_string, new_string)
        
    with open(file_path, 'w') as file:
        file.write(updated_contents)
```

```python
import unicodedata as ud
def upper_greek(str):
    d = {ord('\N{COMBINING ACUTE ACCENT}'):None}
    return ud.normalize('NFD',str).upper().translate(d)
```

```python
import requests

def get_page_content(url):
    response = requests.get(url)
    if response.status_code == 200:
        return response.content
    else:
        print(f"Ошибка при запросе страницы {url}: {response.status_code}")
        return None
```

```python
import os
import re
from bs4 import BeautifulSoup

def clean_text(text):
    pattern = r'[^ \n\u0370-\u03FF\u1F00-\u1FFF.,!?;]'
    cleaned_text = re.sub(pattern, '', text)
    cleaned_text = cleaned_text.replace('  ',' ').replace('\n\n\n','\n\n').replace('\n\n\n','\n\n').replace('\n\n\n','\n\n').replace('\n\n\n','\n\n')

    lines = cleaned_text.split('\n')
    while lines and not any('\u0370' <= char <= '\u03FF' for char in lines[0]):
        lines.pop(0)

    while lines and not any('\u0370' <= char <= '\u03FF' for char in lines[-1]):
        lines.pop()

    return '\n'.join(lines)

def extract_metadata_from_html(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')

    work_info_tag = soup.find('p', class_='WorkInfo')
    if work_info_tag and work_info_tag.find('strong'):
        work_name = work_info_tag.find('strong').get_text()
    else:
        work_name = None

    date_tag = soup.find('a', class_='CanonDat')
    if date_tag:
        date_text = date_tag.get_text()
        cleaned_date_text = clean_date_text(date_text)
        if 'B.C.' in cleaned_date_text:
            century = int(re.search(r'\d+', cleaned_date_text).group())
            period = 'Ancient Greek Literature'
        elif 'A.D.' in cleaned_date_text:
            century = int(re.search(r'\d+', cleaned_date_text).group())
            if century <= 1500:
                period = 'Central Greek Literature'
            else:
                period = 'Modern Greek Literature'
        elif 'Varia' in cleaned_date_text or 'Incertum' in cleaned_date_text:
            period = None
        else:
            period = None
    else:
        period = None

    return work_name, period

def clean_date_text(date_text):
    cleaned_text = re.sub(r'[^0-9A-Za-z\s.-]', '', date_text)
    return cleaned_text

def save_text_to_file(output_dir, work_name, text):
    base_file_path = os.path.join(output_dir, f"{work_name}.txt")
    file_path = base_file_path
    count = 1

    while os.path.exists(file_path):
        with open(file_path, 'r', encoding='utf-8') as existing_file:
            existing_text = existing_file.read()
            if existing_text == text:
                print(f"Файл {file_path[77:]} уже существует.")
                return
        file_path = os.path.join(output_dir, f"{work_name} ({count}).txt")
        count += 1

    while True:
        try:
            with open(file_path, 'w', encoding='utf-8') as output_file:
                output_file.write(text)
            print(f"Текст извлечен и сохранен в файл {file_path[77:]}")
            break
        except OSError as e:
            if len(work_name) > 100:
                work_name = work_name[:len(work_name) // 2]
                file_path = os.path.join(output_dir, f"{work_name}.txt")
                count = 1
            else:
                print(f"Пропуск файла {work_name[77:]}.")
                break

def extract_greek_from_files(root_folder):
    for dirpath, dirnames, filenames in os.walk(root_folder):
        for filename in filenames:
            if re.match(r'^\d{4}_\d{3}\.htm$', filename):
                file_path = os.path.join(dirpath, filename)
                with open(file_path, 'r', encoding='utf-8') as file:
                    html_content = file.read()

                    work_name, period = extract_metadata_from_html(html_content)

                    if work_name and period:
                        cleaned_greek_text = clean_text(html_content)

                        output_dir = os.path.join(root_folder, period)
                        os.makedirs(output_dir, exist_ok=True)
                        save_text_to_file(output_dir, work_name, cleaned_greek_text)
                    elif not work_name:
                        print(f"Название работы не найдено в файле {file_path[77:]}")
                    elif not period:
                        print(f"Период не найден или не определен в файле {file_path[77:]}")
```

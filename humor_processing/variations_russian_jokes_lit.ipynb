```python
import numpy as np
embeddings = np.load('russian_dict.npy', allow_pickle=True).item()

import nltk
from nltk.collocations import BigramCollocationFinder
from nltk.metrics import BigramAssocMeasures
import re

def tokenize_and_find_bigrams(text):
#     text = re.sub(r'\d+', '', text)
    tokens = nltk.wordpunct_tokenize(text)
    finder = BigramCollocationFinder.from_words(tokens)
    return finder.nbest(BigramAssocMeasures().pmi, len(finder.ngram_fd))

def get_embedding(lemma):
    if lemma in embeddings:
        return embeddings[lemma]
    else:
        return None

def get_bigrams_with_embeddings(text):
    bigrams = tokenize_and_find_bigrams(text)
    bigrams_with_embeddings = []
    for bigram in bigrams:
        emb1 = get_embedding(bigram[0])
        emb2 = get_embedding(bigram[1])
        if emb1 is not None and emb2 is not None:
            bigram_vector = np.concatenate((emb1, emb2))
            bigrams_with_embeddings.append(bigram_vector)
    return bigrams_with_embeddings

from scipy.spatial.distance import euclidean
def calculate_euclidean_distances_between_bigrams(bigrams):
    distances = []
    for i in range(len(bigrams) - 1):
        bigram1 = bigrams[i]
        bigram2 = bigrams[i + 1]
        distance = euclidean(bigram1, bigram2)
        distances.append(distance)
    
    return distances

import nltk
import pandas as pd
import numpy as np
import re
import gensim
import matplotlib.pyplot as plt
from sklearn.neighbors import KernelDensity
from sklearn.manifold import TSNE
from gensim.models import Word2Vec
from gensim.models.phrases import Phrases, Phraser
from nltk.collocations import *
import seaborn as sns
from natasha import (
    Segmenter, MorphVocab,
    NewsNERTagger,
    NewsEmbedding,
    NewsMorphTagger,
    Doc
)
import re
import glob
import tqdm

def prepare_russian_text(raw_text):

    segmenter = Segmenter()
    emb = NewsEmbedding()
    morph_tagger = NewsMorphTagger(emb)
    ner_tagger = NewsNERTagger(emb)
    morph_vocab = MorphVocab()

    raw_text = re.sub(r'\d+', '0' , raw_text)

    doc = Doc(raw_text)
    doc.segment(segmenter)
    doc.tag_ner(ner_tagger)
    doc = Doc(raw_text)
    doc.segment(segmenter)
    doc.tag_morph(morph_tagger)

    prepared_text = ''
    for token in doc.tokens:
        if token.pos != 'PUNCT':
            try:
                token.lemmatize(morph_vocab)
                prepared_text += token.lemma.lower()
                prepared_text += ' '
            except Exception as ex:
                prepared_text += token.text.lower()
                prepared_text += ' '
    return prepared_text
```

```python
lemma_jokes = open('lemma_anekdots.txt', 'r').read()
lemma_liter = open('lemma_literature.txt', 'r').read()

bigrams_with_embeddings_jokes = get_bigrams_with_embeddings(lemma_jokes)
distances_series_jokes = calculate_euclidean_distances_between_bigrams(bigrams_with_embeddings_jokes)

bigrams_with_embeddings_liter = get_bigrams_with_embeddings(lemma_liter)
distances_series_liter = calculate_euclidean_distances_between_bigrams(bigrams_with_embeddings_liter)
```

```python
jokes_mean = np.mean(distances_series_jokes) 
liter_mean = np.mean(distances_series_liter)

deviation_jokes = np.square(np.subtract(distances_series_jokes, jokes_mean))
deviation_liter = np.square(np.subtract(distances_series_liter, liter_mean))

variance_jokes = np.sqrt(np.sum(deviation_jokes) / (len(deviation_jokes) - 1))
variance_liter = np.sqrt(np.sum(deviation_liter) / (len(deviation_liter) - 1))

print("Вариация последовательности на юморе:", variance_jokes)
print("Вариация последовательности на литературе:", variance_liter)
```

```python
def generate_z_vectors(distances, pattern):
    z_vectors = []
    for t in range(len(distances) - sum(pattern)):
        z_vector = [distances[t]]
        current_position = t
        for step in pattern:
            current_position += step
            z_vector.append(distances[current_position])
        z_vectors.append(z_vector)
    return z_vectors
```

```python
def split_file_by_size(input_file, max_size_mb=1, output_prefix='output_part_'):
    max_size = max_size_mb * 1024 * 1024
    part_number = 1
    current_size = 0
    output_file = None

    with open(input_file, 'r', encoding='utf-8') as f:
        for line in f:
            line_size = len(line.encode('utf-8'))

            if current_size + line_size > max_size:
                if output_file:
                    output_file.close()
                output_filename = f"{output_prefix}{part_number}.txt"
                print(f"Creating {output_filename}")
                output_file = open(output_filename, 'w', encoding='utf-8')
                part_number += 1
                current_size = line_size

            if not output_file:
                output_filename = f"{output_prefix}1.txt"
                print(f"Creating {output_filename}")
                output_file = open(output_filename, 'w', encoding='utf-8')

            output_file.write(line)
            current_size += line_size

    if output_file:
        output_file.close()

# split_file_by_size('pure_anekdots — копия.txt', 0.1, 'anekdot_part_')
```

```python
from wishart import WishartWHeight

z_vectors_jokes = np.array(z_vectors_jokes)
z_vectors_liter = np.array(z_vectors_liter)
combined_data = np.vstack([z_vectors_jokes, z_vectors_liter])
labels = np.array([0] * len(z_vectors_jokes) + [1] * len(z_vectors_liter))

k = 40
h = 0.05
wish = WishartWHeight(z_vectors_jokes, k, h)
wish.build_clusters()
clusters = wish.ready_clusters
```

```python
from matplotlib import colormaps

cmap = colormaps.get_cmap('tab20')
cluster_colors = cmap(np.linspace(0, 1, len(clusters)))
point_to_cluster = {}
for cluster_id, points in clusters.items():
    for point in points:
        point_to_cluster[point] = cluster_id
```

```python
from sklearn.manifold import TSNE

n_samples = combined_data.shape[0]
perplexity = min(30, n_samples - 1)

tsne = TSNE(n_components=2, perplexity=3, max_iter=1000, random_state=42)
reduced_data = tsne.fit_transform(combined_data)
cmap = colormaps.get_cmap('tab20')
cluster_colors = cmap(np.linspace(0, 1, len(clusters)))
```

```python
jokes_points = []
liter_points = []

jokes_colors = []
liter_colors = []

for i, point in enumerate(reduced_data):
    cluster_id = point_to_cluster.get(i, -1)
    if cluster_id == -1:
        color = 'gray'
    else:
        color = cluster_colors[cluster_id % len(cluster_colors)]

    if labels[i] == 0:
        jokes_points.append(point)
        jokes_colors.append(color)
    elif labels[i] == 1:
        liter_points.append(point)
        liter_colors.append(color)

jokes_points = np.array(jokes_points)
liter_points = np.array(liter_points)
jokes_colors = np.array(jokes_colors)
liter_colors = np.array(liter_colors)

plt.figure(figsize=(12, 8))

if len(jokes_points) > 0:
    plt.scatter(jokes_points[:, 0], jokes_points[:, 1], c='blue', marker='o', label='jokes')

if len(liter_points) > 0:
    plt.scatter(liter_points[:, 0], liter_points[:, 1], c='red', marker='^', label='liter')

# motifs = np.array([np.mean(combined_data[points], axis=0) for points in clusters.values()])
# reduced_motifs = tsne.fit_transform(motifs)
# plt.scatter(reduced_motifs[:, 0], reduced_motifs[:, 1], c='red', marker='x', label='motif')

plt.title('Wishart Clustering')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import calinski_harabasz_score
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
combined_data = scaler.fit_transform(combined_data)

def adjusted_calinski_harabasz(data, labels, T=2):
    ch_score = calinski_harabasz_score(data, labels)

    not_noise_ratio = np.sum(labels != 0) / len(labels)
    
    adjusted_ch = ch_score * (not_noise_ratio ** T)
    
    return adjusted_ch

from sklearn.manifold import TSNE
from utils.clustering.WishartParallelKD import Wishart
import numpy as np

k_values = range(3, 30)
significance_levels = [0.05, 0.1, 0.2, 0.3, 0.4]

best_params = None
best_adjusted_ch = -np.inf

for k in k_values:
    for significance_level in significance_levels:
        wish = Wishart(wishart_neighbors=k, significance_level=significance_level)
        wish.fit(combined_data)
        cluster_labels = wish.object_labels
        
        try:
            adjusted_ch = adjusted_calinski_harabasz(combined_data, cluster_labels, T=2)
        except ValueError:
            continue
        
        if adjusted_ch > best_adjusted_ch:
            best_adjusted_ch = adjusted_ch
            best_params = (k, significance_level)

print(f"Лучшие параметры: k={best_params[0]}, significance_level={best_params[1]}")
print(f"Максимальный скорректированный CH: {best_adjusted_ch}")
```

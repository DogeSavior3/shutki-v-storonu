```python
import numpy as np
embeddings = np.load('../../russian_dict.npy', allow_pickle=True).item()

import nltk
from nltk.collocations import BigramCollocationFinder
from nltk.metrics import BigramAssocMeasures
import re

def tokenize_and_find_bigrams(text):
#     text = re.sub(r'\d+', '', text)
    tokens = nltk.wordpunct_tokenize(text)
    finder = BigramCollocationFinder.from_words(tokens)
    return finder.nbest(BigramAssocMeasures().pmi, len(finder.ngram_fd))

def get_embedding(lemma):
    if lemma in embeddings:
        return embeddings[lemma]
    else:
        return None

def get_bigrams_with_embeddings(text):
    bigrams = tokenize_and_find_bigrams(text)
    bigrams_with_embeddings = []
    for bigram in bigrams:
        emb1 = get_embedding(bigram[0])
        emb2 = get_embedding(bigram[1])
        if emb1 is not None and emb2 is not None:
            bigram_vector = np.concatenate((emb1, emb2))
            bigrams_with_embeddings.append(bigram_vector)
    return bigrams_with_embeddings

from scipy.spatial.distance import euclidean
def calculate_euclidean_distances_between_bigrams(bigrams):
    distances = []
    for i in range(len(bigrams) - 1):
        bigram1 = bigrams[i]
        bigram2 = bigrams[i + 1]
        distance = euclidean(bigram1, bigram2)
        distances.append(distance)
    
    return distances

import nltk
import pandas as pd
import numpy as np
import re
import gensim
import matplotlib.pyplot as plt
from matplotlib import colormaps
from sklearn.neighbors import KernelDensity
from sklearn.manifold import TSNE
from gensim.models import Word2Vec
from gensim.models.phrases import Phrases, Phraser
from nltk.collocations import *
import seaborn as sns
from natasha import (
    Segmenter, MorphVocab,
    NewsNERTagger,
    NewsEmbedding,
    NewsMorphTagger,
    Doc
)
import re
import glob
import tqdm

def prepare_russian_text(raw_text):

    segmenter = Segmenter()
    emb = NewsEmbedding()
    morph_tagger = NewsMorphTagger(emb)
    ner_tagger = NewsNERTagger(emb)
    morph_vocab = MorphVocab()

    raw_text = re.sub(r'\d+', '0' , raw_text)

    doc = Doc(raw_text)
    doc.segment(segmenter)
    doc.tag_ner(ner_tagger)
    doc = Doc(raw_text)
    doc.segment(segmenter)
    doc.tag_morph(morph_tagger)

    prepared_text = ''
    for token in doc.tokens:
        if token.pos != 'PUNCT':
            try:
                token.lemmatize(morph_vocab)
                prepared_text += token.lemma.lower()
                prepared_text += ' '
            except Exception as ex:
                prepared_text += token.text.lower()
                prepared_text += ' '
    return prepared_text
```

```python
lemma_jokes = open('../../lemma_anekdots.txt', 'r').read()[0:10000]
lemma_liter = open('../../lemma_literature.txt', 'r').read()[0:10000]

bigrams_with_embeddings_jokes = get_bigrams_with_embeddings(lemma_jokes)
distances_series_jokes = calculate_euclidean_distances_between_bigrams(bigrams_with_embeddings_jokes)

bigrams_with_embeddings_liter = get_bigrams_with_embeddings(lemma_liter)
distances_series_liter = calculate_euclidean_distances_between_bigrams(bigrams_with_embeddings_liter)

# jokes_mean = np.mean(distances_series_jokes) 
# liter_mean = np.mean(distances_series_liter)

# deviation_jokes = np.square(np.subtract(distances_series_jokes, jokes_mean))
# deviation_liter = np.square(np.subtract(distances_series_liter, liter_mean))

# variance_jokes = np.sqrt(np.sum(deviation_jokes) / (len(deviation_jokes) - 1))
# variance_liter = np.sqrt(np.sum(deviation_liter) / (len(deviation_liter) - 1))

# print("Вариация последовательности на юморе:", variance_jokes)
# print("Вариация последовательности на литературе:", variance_liter)

# anekdots_splitted = [i for i in open('../../pure_anekdots.txt', 'r', encoding = 'utf-8').read().split('\n\n') if i != '']
# def count_variance(distances):
#     deviation = np.square(np.subtract(distances, np.mean(distances)))
#     return np.sqrt(np.sum(deviation) / (len(deviation) - 1))

# variance_series_for_jokes = [count_variance(calculate_euclidean_distances_between_bigrams(get_bigrams_with_embeddings(prepare_russian_text(i)))) for i in anekdots_splitted[0:200]]

# def split_text_into_chunks(text, chunk_size):
#     words = text.split()
#     chunks = [' '.join(words[i:i + chunk_size]) for i in range(0, len(words), chunk_size)]
#     return chunks

# chunks = split_text_into_chunks(open('../../pure_literature.txt', 'r', encoding = 'utf-8').read(), 25)

# variance_series_for_liter = [count_variance(calculate_euclidean_distances_between_bigrams(get_bigrams_with_embeddings(prepare_russian_text(i)))) for i in chunks[0:200]]

def generate_z_vectors(distances, pattern):
    z_vectors = []
    for t in range(len(distances) - sum(pattern)):
        z_vector = [distances[t]]
        current_position = t
        for step in pattern:
            current_position += step
            z_vector.append(distances[current_position])
        z_vectors.append(z_vector)
    return z_vectors

pattern = [1, 1, 1, 1, 1, 1, 1, 1, 1]
z_vectors_jokes = generate_z_vectors(distances_series_jokes, pattern)
z_vectors_liter = generate_z_vectors(distances_series_liter, pattern)
```

```python
from WishartParallelKD import Wishart

combined_data = np.vstack([z_vectors_jokes, z_vectors_liter])
labels = np.array([0] * len(z_vectors_jokes) + [1] * len(z_vectors_liter))

k = 10
significance_level = 0.2

wish = Wishart(wishart_neighbors=k, significance_level=significance_level)

wish.fit(combined_data)
cluster_labels = wish.object_labels

tsne = TSNE(n_components=2, perplexity=min(30, len(combined_data)-1), random_state=42)
reduced_data = tsne.fit_transform(combined_data)

unique_clusters = np.unique(cluster_labels)
cmap = colormaps.get_cmap('tab20')
cluster_colors = cmap(np.linspace(0, 1, len(unique_clusters)))

plt.figure(figsize=(12, 8))
for cluster_id in unique_clusters:
    cluster_points = reduced_data[cluster_labels == cluster_id]
    color = cluster_colors[cluster_id % len(cluster_colors)]
    
    if cluster_id == 0:
        plt.scatter(cluster_points[:, 0], cluster_points[:, 1], c='gray', marker='.', label='Noise')
    else:
        marker = 'o' if cluster_id <= len(z_vectors_jokes) else '^'
        label = 'joke' if cluster_id <= len(z_vectors_jokes) else 'liter'
        plt.scatter(cluster_points[:, 0], cluster_points[:, 1], c=color, marker=marker, label=label)

plt.title('Wishart Clustering')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.legend()
plt.grid(True)
plt.show()

for cluster_id in unique_clusters:
    if cluster_id == 0:
        print(f"Cluster {cluster_id} (Noise): {np.sum(cluster_labels == cluster_id)} points")
    else:
        joke_count = np.sum((cluster_labels == cluster_id) & (labels == 0))
        liter_count = np.sum((cluster_labels == cluster_id) & (labels == 1))
        print(f"Cluster {cluster_id}: jokes={joke_count}, literature={liter_count}")
```
